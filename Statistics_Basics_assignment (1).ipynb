1.  Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.
--> Data can be classified into two main types: qualitative and quantitative.

a. Qualitative Data (Categorical Data):

Definition: Describes qualities or characteristics. This data cannot be measured numerically.

Examples:

Colors of cars (red, blue, green)

Types of fruits (apple, banana, orange)

Gender (male, female, non-binary)

Yes/No responses

b. Quantitative Data (Numerical Data):

Definition: Describes quantities. This data can be measured and expressed numerically.

Examples:

Age (21, 35, 70)

Height (5.9 feet, 180 cm)

Number of students in a class (25)

Temperature (22°C)

Scales of Measurement:

There are four levels of measurement that further classify data: Nominal, Ordinal, Interval, and Ratio. These apply to both qualitative and quantitative data.

a. Nominal Scale:

Type: Qualitative

Definition: Categories without any order.

Key Feature: Labels or names only; no mathematical meaning.

Examples:

Blood types (A, B, AB, O)

Nationality (American, Canadian, Indian)

Marital status (Single, Married, Divorced)

b.  Ordinal Scale:

Type: Qualitative (sometimes quantitative in practice)

Definition: Categories with a meaningful order, but differences between them aren’t uniform.

Key Feature: Ranked data, but we can’t do precise arithmetic.

Examples:

Satisfaction ratings (Poor, Fair, Good, Excellent)

Education level (High School, Bachelor’s, Master’s, PhD)

Class rankings (1st, 2nd, 3rd)

c. Interval Scale:

Type: Quantitative

Definition: Numeric scales with equal intervals between values, but no true zero.

Key Feature: We can add and subtract, but ratios don’t make sense.

Examples:

Temperature in Celsius or Fahrenheit

IQ scores

Calendar years (1990, 2000, 2020)

d.  Ratio Scale:

Type: Quantitative

Definition: Numeric scale with equal intervals and a true zero.

Key Feature: Supports all arithmetic operations, including ratios.

Examples:

Height, weight

Age (e.g., 20 years is twice as old as 10 years)

Income, distance, number of children

2.  What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.
--> Measures of central tendency are ways to describe the center or typical value in a dataset. The three main measures are mean, median, and mode.

a. Mean (Average):

Definition:
The mean is the sum of all values divided by the number of values.

Formula:

Mean = Sum of all values/Number of values

Example:
Test scores: 70, 80, 90

Mean = 70 + 80 + 90/3 = 80

Best Use:

When data is normally distributed (no extreme outliers).

Useful when every value matters.

b. Median:

Definition:
The middle value when data is ordered. If there's an even number of values, take the average of the two middle ones

Example:

Incomes: $30k, $35k, $40k, $45k, $5M

Ordered: $30k, $35k, $40k, $45k, $5M

Median = $40k

Best Use:

When data has outliers or is skewed.

Use with income, house prices, or any data with extreme highs/lows.

c. Mode

Definition:
The most frequently occurring value(s) in a dataset. There can be no mode, one mode, or multiple modes.

Example:

Favorite colors: red, blue, blue, green, red, blue

Mode = blue

Best Use:

When you want to find the most common category or value.

Great for categorical data (e.g., most popular product, most common shoe size).

3.  Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?
--> Dispersion refers to how spread out or scattered the values in a dataset are. It tells you how much the values differ from the central tendency (like the mean).

Importance of dispersion

Two datasets can have the same mean but very different spreads.

Dispersion helps you understand the consistency or variability in the data.

a. Variance:

Definition:
Variance measures the average of the squared differences from the mean.

Interpretation:

A larger variance means data is more spread out.

A smaller variance means data is clustered closer to the mean.

Example:

Data: 4, 5, 6

Mean = 5

Deviations: -1, 0, +1 → Squares: 1, 0, 1

Variance = (1 + 0 + 1) / 3 = 0.67

b. Standard Deviation (SD)

Definition:
The square root of the variance. It brings the units back to the original data scale.

Use:

Easier to interpret than variance because it’s in the same units as the data.

Widely used in statistics, science, finance, etc.

Example (continued):
Variance = 0.67
Standard Deviation = √0.67 ≈ 0.82

4.  What is a box plot, and what can it tell you about the distribution of data?
--> A box plot (also called a box-and-whisker plot) is a graphical summary of a dataset that shows its distribution, central value, and variability. It's super useful for spotting outliers and understanding how data is spread out.

A box plot visually displays five key summary statistics:

Minimum – the smallest value (excluding outliers)

First Quartile (Q1) – the 25th percentile

Median (Q2) – the 50th percentile (middle of the data)

Third Quartile (Q3) – the 75th percentile

Maximum – the largest value (excluding outliers)

What Can a Box Plot Tell You?

 a. Center of the Data

The line inside the box is the median — the middle value.

It gives you a quick idea of the central tendency.

b. Spread of the Data

The width of the box shows the interquartile range (IQR), which is Q3 - Q1.

The IQR captures the middle 50% of the data.

c. Skewness (Symmetry)

If the median is closer to the bottom or top of the box, the data is skewed.

Median closer to Q1 → Right (positive) skew

Median closer to Q3 → Left (negative) skew

d. Outliers

Points that fall far outside the whiskers (typically 1.5 × IQR from Q1 or Q3) are marked as individual dots or stars.

5.  Discuss the role of random sampling in making inferences about populations.
--> Random sampling is a technique where every member of a population has an equal chance of being selected to be part of the sample.

When we want to learn something about a population, it's usually too large to study every individual. So we take a sample, study it, and use the results to make inferences (educated guesses) about the whole group.

Random sampling ensures that:

The sample is representative of the population.

Results are generalizable beyond just the sample.

Bias is minimized, making the conclusions more reliable.

6.  Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
--> Skewness measures the asymmetry of a dataset's distribution around its mean.

If a distribution is perfectly symmetrical, it has zero skewness.

If it’s not symmetrical, it’s skewed — either to the left or right.

Types of Skewness

a. Symmetrical Distribution (Zero Skewness)

The mean, median, and mode are equal or very close.

The left and right sides of the distribution are mirror images.

 Example: Heights of adults in a large population.

b. Positive Skew (Right Skew)

The tail on the right side is longer.

Most data values are clustered on the left, but a few large values stretch the tail to the right.

Mean > Median > Mode

 Example: Income data — most people earn a moderate amount, but a few earn very high salaries.

c. Negative Skew (Left Skew)

The tail on the left side is longer.

Most data values are clustered on the right, but a few small values stretch the tail to the left.

Mean < Median < Mode

 Example: Age at retirement — most retire around 65, but some may retire very early (e.g., at 40), pulling the mean down.

How Skewness Affects Data Interpretation

Choosing the right measure of central tendency: Use median instead of mean when data is skewed.

Understanding real-world phenomena: Skewness tells you if there are extreme values affecting results.

Making predictions or decisions: You don’t want to base decisions on a misleading average.

--> The Interquartile Range (IQR) is a measure of spread that describes the range of the middle 50% of a dataset.

Outliers are values that fall much lower or higher than most of the data.

a. Outlier Rule:

A common method defines outliers as values that fall:

Below:

𝑄1 − 1.5 × IQR

Above:

Q3+1.5×IQR

Any values outside this range are considered potential outliers.

b. IQR in Box Plots

In a box plot:

The box spans from Q1 to Q3 (IQR)

The whiskers extend to the lowest and highest values within the outlier range

Any dots beyond the whiskers = outliers

8. . Discuss the conditions under which the binomial distribution is used.
-->Conditions for Using the Binomial Distribution:

To use the binomial distribution properly, all of the following must be true:

a. Fixed Number of Trials (n)

The number of experiments or observations is set in advance.

Example: You flip a coin 20 times — that’s a fixed number.

b. Two Possible Outcomes (Success/Failure)

Each trial must result in one of two outcomes:

 Success

 Failure

Example: Rolling a die and calling a “success” when you roll a 6.

c. Constant Probability of Success (p)

The probability of success stays the same for each trial.

Example: If you're flipping a fair coin, the chance of heads (success) is always 0.5.

d. Independent Trials
The outcome of one trial does not affect the outcome of another.

Example: Drawing cards with replacement is independent. Without replacement? Not binomial.

9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).
--> The normal distribution is a continuous, symmetric probability distribution that describes how data tends to be distributed around a mean.

Key Properties:

a. Symmetrical Shape

The curve is perfectly symmetrical about the mean.

Left and right sides are mirror images.

b. Bell-Shaped Curve

Most data clusters around the mean, with fewer values far from it.

c. Mean = Median = Mode

All measures of central tendency are equal and located at the center.

d. Defined by Two Parameters:

Mean (μ) – the center of the distribution

Standard Deviation (σ) – controls the spread

e. Tails Extend Infinitely

Never touch the x-axis but approach it asymptotically.

11.  Explain what a random variable is and differentiate between discrete and continuous random variables.

A random variable is a numerical outcome of a random process or experiment.

It's a way to assign numbers to outcomes of a random event.

Think of it as a function that maps outcomes to numbers

a. Discrete Random Variable

Takes on a finite or countable number of values.

Each possible value has a specific probability.

Often used when counting things (e.g., number of kids, number of cars, dice rolls).

b.  Continuous Random Variable

Can take on any value within a range or interval.

Values are uncountably infinite.

Often used when measuring things (e.g., time, height, weight, temperature).
